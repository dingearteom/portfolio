```python
sigma_noise = 0.2

X = np.random.uniform(0, 10, size=(10))
y = np.sin(X) + np.random.normal(0, sigma_noise ** 2, X.shape[0])

a = list(zip(X, y))
a = sorted(a)
X, y = zip(*a)
X = np.array(X)[:, np.newaxis]
y = np.array(y)

gr = GaussianProcessRegression(noisy=False)
gr.fit(X, y)
plot(gr, [(0, 10)], ax=plt)
```
