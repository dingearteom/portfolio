```python
from BO.bo import BO
import numpy as np
from GP.plotting import plot
import matplotlib.pyplot as plt
plt.style.use('classic')
```

```python
def f(x):
    return np.sin(x[0][0]) * np.sin(x[0][1]) 
```

```python
my_model = BO(lambda x: np.sin(x[0]) * np.sin(x[1]), [(0, 2 * np.pi), (0, 2 * np.pi)])
```

```python
my_model.run_optimization(100)
```

```python
plot(my_model.gr, [(0, 2 * np.pi), (0, 2 * np.pi)])
```

```python
bounds = [{'name':'var1', 'type':'continuous', 'domain':(0, 2 * np.pi)}, 
          {'name':'var2', 'type':'continuous', 'domain':(0, 2 * np.pi)}]
```

```python
model = GPyOpt.methods.BayesianOptimization(f=f,
                                             domain=bounds[:2],
                                             model_type='GP',
                                             acquisition_type='EI',
                                             normalize_Y=True,
                                             acquisition_weight=2,
                                             model_update_interval=1,
                                             verbosity=True)
```

```python
model.run_optimization(100)
```
