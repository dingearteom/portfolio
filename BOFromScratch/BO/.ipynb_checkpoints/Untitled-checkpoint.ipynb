```python
from scipy.special import erfc, erf
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(-3, 3)
plt.plot(x, erf(x))
plt.xlabel('$x$')
plt.ylabel('$erfc(x)$')
plt.show()
```
