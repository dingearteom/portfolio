```python
sigma_noise = 0

X = np.random.uniform(0, 10, size=(3))
y = np.sin(X) + np.random.normal(0, sigma_noise ** 2, X.shape[0])
X = X[:, np.newaxis]

gr = GaussianProcessRegression(noisy=False)
gr.fit(X, y)
plot(gr, [(0, 10)], ax=plt)
```

```python
sigma_noise = 0.05
X = np.random.uniform(-3.,3.,(50,2))
Y = np.sin(X[:,0:1]) * np.sin(X[:,1:2])+np.random.randn(50,1)*sigma_noise
Y = Y.flatten()
gr = GaussianProcessRegression(noisy=True)
gr.fit(X, Y)
plot(gr, [(-3, 3), (-3, 3)], ax=plt)
```
